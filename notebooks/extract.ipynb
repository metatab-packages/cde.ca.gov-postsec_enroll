```python
import seaborn as sns
import metapack as mp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display 

%matplotlib inline
sns.set_context('notebook')
mp.jupyter.init()
```

```python
#pkg = mp.jupyter.open_package()
pkg = mp.jupyter.open_source_package()
pkg
```

```python
cgr12 = pd.concat([ r.dataframe(parse_dates=False) for r in pkg.references() if r.group == '12' ])
cgr12.head().T
```

```python
cgr16 = pd.concat([ r.dataframe() for r in pkg.references() if r.group == '16' ])
cgr16.head().T
```

```python
common_cols =  [c for c in cgr12.columns if c in set(cgr12) & set(cgr16)] # Comprehension in order to get same order as in dataframe
cols12 = [c for c in cgr12.columns if c in set(cgr12) ^ set(cgr16)]
cols16 = [c for c in cgr16.columns if c in set(cgr12) ^ set(cgr16)]
cols16
```

```python
t = cgr12.merge(cgr16, on=common_cols, how='outer')
```

```python
len(cgr12), len(cgr16), len(t)
```

```python
t = t.rename(columns={list(t.columns)[0]:'AcademicYear'}) # It's got unicode crap in the name
def repl(v):
    vo = v
    v = v.replace('(16 Months)','16m')
    v = v.replace('(12 Months)','12m')
    v = v.replace('In-State','In')
    v = v.replace('Out-of-State','Out')
 
    return v
    
[repl(c) for c in t.columns]
```

```python
cgr = t
```

```python
for c in cgr.columns:
    try:
        cgr[c] = pd.to_numeric(cgr[c].replace('*','NaN').replace('NaN',np.nan) )
        cgr[c] = cgr[c].astype('Int64')
        
        try:
            cgr[c] = cgr[c].astype('int')
        except Exception as e:
            cgr[c] = cgr[c].astype('float') # Has missing values, so can't be an int
        
    except ValueError:
        print(f"Failed to convert to numeric: {c}")
    except TypeError:
        print(f"Failed to convert to integer: {c}")
```
